```python
from keras.datasets import mnist
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
from keras import backend as K
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
train_datagen = ImageDataGenerator(rotation_range=60)
```

```python
train_datagen.fit(X_train)
```

```python
for X_batch, y_batch in train_datagen.flow(X_train, y_train, batch_size=9):
    for i in range(0, 9):
        pyplot.subplot(330 + 1 + i)
        pyplot.imshow(X_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))# show the plot
    pyplot.show()
    break
```

```python
## random shearing and zooming for data augmentation
```

```python
train_datagen_shear_zoom = ImageDataGenerator(shear_range=0.5,
                                             zoom_range=0.5)
train_datagen_shear_zoom.fit(X_train)
```

```python
for X_batch, y_batch in train_datagen_shear_zoom.flow(X_train, y_train, batch_size=9):
    for i in range(0, 9):
        pyplot.subplot(330 + 1 + i)
        pyplot.imshow(X_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))# show the plot
    pyplot.show()
    break
```

```python
## horizontal and vertical flips for data augmentation
```

```python
train_datagen_vertical_horizontal = ImageDataGenerator(vertical_flip=True,
                                                      horizontal_flip=True)
train_datagen_vertical_horizontal.fit(X_train)
```

```python
for X_batch, y_batch in train_datagen_vertical_horizontal.flow(X_train, y_train, batch_size=9):
    for i in range(0, 9):
        pyplot.subplot(330 + 1 + i)
        pyplot.imshow(X_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))# show the plot
    pyplot.show()
    break
```

```python
## clearing flips not a good idea in training images for digit recognition
```

```python
## random shift for data augmentation
```

```python
train_datagen_shift = ImageDataGenerator(width_shift_range=0.3,
                                        height_shift_range=0.3)

train_datagen_shift.fit(X_train)
```

```python
for X_batch, y_batch in train_datagen_shift.flow(X_train, y_train, batch_size=9):
    for i in range(0, 9):
        pyplot.subplot(330 + 1 + i)
        pyplot.imshow(X_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))# show the plot
    pyplot.show()
    break
```

```python
## applying all of the above for data augmentaion
```

```python
train_datagen_all = ImageDataGenerator(
                    rotation_range=45,
                    width_shift_range=0.2,
                    height_shift_range=0.2,
                    shear_range=0.2,
                    zoom_range=0.2,
                    horizontal_flip=True,
                    fill_mode='nearest')
train_datagen_all.fit(X_train)
```

```python
for X_batch, y_batch in train_datagen_all.flow(X_train, y_train, batch_size=9):
    for i in range(0, 9):
        pyplot.subplot(330 + 1 + i)
        pyplot.imshow(X_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))# show the plot
    pyplot.show()
    break
```
